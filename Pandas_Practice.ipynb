**8-Solucione o problema das colunas "Value" e "Weight" transformando-as de strings para inteiros**

Considere esse exercício como um desafio, se não conseguir, tente buscar na internet, como em sites e fóruns de programação. O "know how" em pesquisar esses tipos de coisas é extremamente importante em qualquer área relacionada a códigos.    

```python
#problema do Weight
jogadores['Weight'] = jogadores['Weight'].apply(lambda x : float(str(x).replace('lbs','')))


#Problema de Value
def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)

jogadores['Value'] = jogadores['Value'].apply(lambda x: extract_value_from(x))
```

## Parte II
### Filtrando e explorando DataFrames

**1-Qual é o jogador mais velho? E o mais novo?**

```python
jogadores.head()
```

```python
#PRIMEIRA MANEIRA

#Jogador mais velho
jogadores[jogadores['Age']==jogadores['Age'].max()]['Name']

#Jogador mais novo
jogadores[jogadores['Age']==jogadores['Age'].min()]['Name']

#SEGUNDA MANEIRA

#Jogador mais velho
print(jogadores.sort_values('Age').tail(1))

#Jogador mais novo
print(jogadores.sort_values('Age').head(1))
```

**3-O Brasil é o país do futebol... mas será? Qual o top 5 países que mais tem jogadores no Fifa 19?**
