```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('http://coreyms.com').text
```

```python
soup=BeautifulSoup(source,'lxml')
```

```python
print(soup.prettify())
```

```python
article=soup.find('article')
#print(article.prettify())
```

```python
headline=article.h2.a.text
print(headline)
```

```python
summary=article.find('div',class_='entry-content').p
print(summary)
```

```python
summary=article.find('div',class_='entry-content').p.text
print(summary)
```

```python
video=article.find('iframe',class_='youtube-player')
print(video)
```

```python
video=article.find('iframe',class_='youtube-player')['src']
print(video)
```

```python
video_id=video.split('/')[4].split('?')[0]
print(video_id)
```

## Aplicando isso em todos os Títulos 

```python
for article in soup.find_all('article'):
    headline=article.h2.a.text
    print(headline)
    
    summary=article.find('div',class_='entry-content').p.text
    print(summary)
    
    
    print()
   
```
